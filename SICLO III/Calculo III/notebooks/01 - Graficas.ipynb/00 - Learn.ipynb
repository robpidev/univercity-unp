# Gráficas

```julia
using Plots
```

## 1. Gráficas en dos dimeciones

```julia
x = -pi:0.01:pi
y = sin.(x)

plot(x, y, label = "sin(x)", framestyle=:origin )
```

## 2. Layout

```julia

plot(x, y, layout=(2,1), label = "sin(x)")
```

## 3. Plot3D

```julia
t = -pi:pi/200:4pi
x = cos.(t)
y = sin.(t)
plot3d(x, y, t, framestyle=:origin )
```

## 4. Superficie

```julia
x = -1:0.1:1
y = -1:0.1:1
h(x, y) = x^2 - y^2

surface(x, y, h)
```

## 5. Malla

```julia
wireframe(x, y, h, framestyle=:origin)
```

## 6. Animations

```julia
anim = @animate for i=1:100
    x = 0:0.01:2i*pi/100
    plot(sin.(x), cos.(x), xlim=(-1.2, 1.2), ylim=(-1.2, 1.2), aspect_ratio=1, framestyle=:origin)
end

gif(anim, fps=30)
```
